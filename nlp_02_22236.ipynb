# implementing tr commands using regular expression.
the commands done are for finding individual words and then finding the frequency of each word.

```python
x_tok=re.findall("\\b[^ ][A-Za-z][^\:\$\#\!\@\%\^\&\*\(\)\_\-\+\=]*?\\b",x)
x_tok
```

```python
print(''in x_tok)
```

```python
x_tok.sort(key=len)
print(f"the sorted list is:\n{x_tok}")
```

```python
def rank(tok,word):
    num=0
    for w in tok:
        if w==word:
            num+=1
    return num,word
```

```python
ranked_x=[]
for i in range(len(x_tok)):
    if x_tok[i] in ranked_x:
        continue
    else:
        ranked_x.append(rank(x_tok,x_tok[i]))
ranked_x_final=list(set(ranked_x))
```
